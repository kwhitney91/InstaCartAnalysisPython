```python
# Aggregating data using the transform function
df_ords_prod_merge['max_order'] = df_ords_prod_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prod_merge['max_order'] = df_ords_prod_merge.groupby(['user_id'])['order_number'].transform("max")
```

```python
df_ords_prod_merge.head(15)
```

```python
#Creating a loyalty Flag
df_ords_prod_merge.loc[df_ords_prod_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prod_merge.loc[(df_ords_prod_merge['max_order'] <= 40) & (df_ords_prod_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prod_merge.loc[df_ords_prod_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
