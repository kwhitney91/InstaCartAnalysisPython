#Question 4- create a loyalty flag for existing customers using the transform() and loc() functions.

```python
#use groupby and transform to create loylaty flag
df_ords_prod_merge['max_order'] = df_ords_prod_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#updated to aviod warning code
df_ords_prod_merge['max_order'] = df_ords_prod_merge.groupby(['user_id'])['order_number'].transform("max")
```

```python
#check updated data frame
df_ords_prod_merge.head(25)
```

```python
#Creating a loyalty Flag
df_ords_prod_merge.loc[df_ords_prod_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prod_merge.loc[(df_ords_prod_merge['max_order'] <= 40) & (df_ords_prod_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prod_merge.loc[df_ords_prod_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flag
df_ords_prod_merge['loyalty_flag'].value_counts(dropna = False)
```

# Question 5- Look at spending habits of differnt groups
Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).

```python
#create an agg()function to ccheck spending habits based on loyalty_flag
df_ords_prod_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

This shows that the price of an item purchased by a new customer is about $3 more expensive then an item purchased by a loyal customer. 

# Question 6 
Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# use transfor to find mean product price
df_ords_prod_merge['mean_product_price'] = df_ords_prod_merge.groupby(['user_id'])['prices'].transform("mean")
```

```python
# check relevent columns
df_ords_prod_merge[['user_id','mean_product_price']].head(25)
```

```python
#create spending flag column
df_ords_prod_merge.loc[df_ords_prod_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prod_merge.loc[df_ords_prod_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#check frequencey
df_ords_prod_merge['spending_flag'].value_counts(dropna = False)
```

```python
#check new created columns of dataframe
df_ords_prod_merge[['user_id', 'mean_product_price', 'spending_flag']].head(25)
```

# Question 7
Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# use transform to find medium days between order for each person
df_ords_prod_merge['median_days_between_orders'] = df_ords_prod_merge.groupby(['user_id'])['days_since_prior_order'].transform("median")
```

```python
# check relevent columns
df_ords_prod_merge[['user_id','median_days_between_orders','days_since_prior_order']].head(25)
```

```python
#Create flag for order frequency
df_ords_prod_merge.loc[df_ords_prod_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent_customer'
```

```python
df_ords_prod_merge.loc[df_ords_prod_merge['median_days_between_orders'] <= 20 & (df_ords_prod_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular_customer'
```

```python
df_ords_prod_merge.loc[df_ords_prod_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent_customer'
```

```python
# check value counts
df_ords_prod_merge['order_frequency'].value_counts(dropna = False)
```

Seems like there are a good amount of customers that do not a value assigned, this is most likley because these are the users who have only made one order

```python
# Check relevent columns
df_ords_prod_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency']].head(25)
```
